```python
def format(index):
    elem = [int(s) for s in df['pixels'][index].split(' ')]
    return elem
```

## Re-fromatting the pixel data ## 

##### Using a for loop to do it #####
This is markdown

```python
arr = []

for i in tqdm(range(35887)):
    arr.append(format(i))
```

```python
df['pixel_formatted'] = arr
```
